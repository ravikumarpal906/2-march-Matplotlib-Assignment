Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

**What is Matplotlib?**

Matplotlib is a comprehensive Python library for creating static, animated, and interactive visualizations in Python. It provides a MATLAB-like plotting framework and is designed to generate high-quality figures in various formats, including PNG, JPG, PDF, SVG, and EPS.

**Why is Matplotlib used?**

Matplotlib is widely used for various reasons:

1. **Extensive Plotting Capabilities:** It offers a wide range of plot types, including line plots, scatter plots, bar plots, histograms, pie charts, and 3D plots.
2. **Customization:** Matplotlib provides extensive customization options for axes, labels, titles, legends, colors, and styles, enabling precise control over the appearance of plots.
3. **Integration with Other Libraries:** Matplotlib seamlessly integrates with other popular Python libraries such as NumPy, SciPy, and Pandas, making it convenient for data analysis and visualization tasks.
4. **Cross-Platform Support:** Matplotlib supports multiple operating systems, including Windows, macOS, and Linux, ensuring portability and accessibility.
5. **Interactive Plotting:** With the `pyplot` interface, Matplotlib allows for interactive plotting, enabling users to zoom, pan, and explore data visually.

**Five Plots that can be plotted using the Pyplot module of Matplotlib:**

1. **Line Plot:** Displays data points connected by straight lines, useful for visualizing trends and relationships between variables.
2. **Scatter Plot:** Plots individual data points as markers, helpful for visualizing the distribution of data and identifying patterns or clusters.
3. **Bar Plot:** Represents data as rectangular bars, suitable for comparing categorical data or showing frequencies.
4. **Histogram:** Displays the distribution of data by dividing it into bins and plotting the number of data points in each bin.
5. **Pie Chart:** Represents data as slices of a pie, illustrating the proportions of different categories within a dataset.


Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data

**Scatter Plot:**

A scatter plot is a type of plot that displays individual data points as markers on a coordinate plane. It is commonly used to visualize the relationship between two variables, with one variable represented on the x-axis and the other on the y-axis. Scatter plots are particularly useful for identifying patterns, clusters, and trends in data.

**Code to Generate Data for x and y:**

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

import numpy as np

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])    
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

The subplot() function in matplotlib is used to create multiple plots in a single figure. It’s very useful when you want to visualize different categories of data side by side for comparison.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(10, 6))

#Line 1
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line 1')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

#Line 2
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line 2')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Line 3
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line 3')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Line 4
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line 4')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


**Bar Plot:**

A bar plot is a graphical representation of data where the length of a bar represents the value of the data point. Bar plots are commonly used to compare different categories or groups of data. They are particularly useful for visualizing categorical data, such as the number of students in different classes or the sales of different products in different regions.

**Why is it used?**

Bar plots are used for various reasons:

- **Comparison:** Bar plots enable easy comparison of different categories or groups of data. The lengths of the bars visually represent the relative values, making it easy to identify which categories have higher or lower values.
- **Categorical Data:** Bar plots are ideal for visualizing categorical data, where the categories are represented on the x-axis and the values are represented by the bar lengths.
- **Frequency Distribution:** Bar plots can be used to visualize the frequency distribution of data. For example, a bar plot can be used to show the number of students in different age groups or the number of sales in different months.
- **Time Series Data:** Bar plots can also be used to visualize time series data, where the categories represent different time periods and the bar lengths represent the values for each period.

**Example:**


```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Bar Plot')
plt.show()

plt.barh(company, profit)
plt.xlabel('Profit')
plt.ylabel('Company')
plt.title('Horizontal Bar Plot')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.


**Box Plot:**

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of data. It visually displays the median, quartiles, and outliers of the data set. Box plots are particularly useful for comparing the distributions of different groups of data or for identifying outliers.

**Why is it used?**

Box plots are used for various reasons:

- **Distribution Visualization:** Box plots provide a quick and informative overview of the distribution of data. They show the spread of the data, the median, and the presence of any outliers.
- **Comparison of Distributions:** Box plots enable easy comparison of the distributions of different groups of data. By placing multiple box plots side-by-side, it is possible to visually assess the similarities and differences between the groups.
- **Outlier Detection:** Box plots are effective in identifying outliers, which are data points that significantly deviate from the rest of the data. Outliers can be valuable for further investigation or analysis.
- **Data Cleaning:** Box plots can be used as a data cleaning tool. Outliers identified in box plots can be further investigated and potentially removed if they are considered erroneous or irrelevant.

**Example:**

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1, box2])
plt.show()
```
